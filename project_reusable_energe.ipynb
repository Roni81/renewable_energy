```python
import pandas as pd
import numpy as np
from pandas import Series
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc("font", family = "Malgun Gothic")
sns.set(font="Malgun Gothic", 
rc={"axes.unicode_minus":False}, style='white')
```

```python
sns.__version__
```

```python
plant_sts = pd.read_csv("C:/Users/user/reusable_enserge_stat_Kyunggee.csv",encoding = 'euc-kr')

plant_sts[plant_sts.plant == '태양광'].loc[:, 'count']
kind_plant = plant_sts.groupby("plant")['count'].sum()
kind_plant_df = pd.DataFrame(kind_plant)
kind_plant_df = kind_plant_df.rename(columns = {'count':'amount'})
kind_plant_df = kind_plant_df.reset_index()
```

```python
kind_plant_df
```

```python

#sns.barplot(x = kind_plant_df.index, y = kind_plant_df['amount'])

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
fig.subplots_adjust(hspace = 0.05)

ax = sns.barplot(x = 'plant', y= "amount", data = kind_plant_df, ax = ax1)
ax1.set_xlabel('', fontsize = 16)
ax = sns.barplot(x = 'plant', y= "amount", data = kind_plant_df, ax = ax2)
# ax1.plot(x = kind_plant_df['plant'])
# ax2.plot(x = kind_plant_df['plant'])
ax1.set_ylim(30000,30500)
ax2.set_ylim(0, 30)

ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)

ax1.yaxis.grid()
ax2.yaxis.grid()

kwargs = dict(marker=[(-1, -0.5), (1, 0.5)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)

```
