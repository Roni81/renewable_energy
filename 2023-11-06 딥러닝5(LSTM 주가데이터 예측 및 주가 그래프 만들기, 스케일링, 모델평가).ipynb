# 연도별로 원하는 기간 설정하기

```python
ts=[['2000', '2010'],
    ['2010', '2015'],
    ['2015', '2020'],
    ['2020', '2023']]
```

```python
ts
```

### 2행 2열로 나누고 fig와 axes가 리턴

```python
fig, axes=plt.subplots(2,2) # 2행 2열
fig.set_size_inches(16,9)
for i in range(4): # 4번 반복을 통해 그래프 생성
    ax = axes[i//2, i%2] # 2로 나누어서 0 혹은 1이 나오면 i가 0이면 0,0, i가 1이면 0,1...이런식
    df = kakao.loc[(kakao.index > ts[i][0]) & (kakao.index <ts[i][1])] # 
    sns.lineplot(y=df['Close'], x=df.index, ax=ax)
    ax.set_title(f'{ts[i][0]}~{ts[i][1]}')
    ax.set_xlabel('time')
    ax.set_ylabel('price')
```

```python
fig, axes=plt.subplots(2,2) # 2행 2열
fig.set_size_inches(16,9)
for i in range(4): # 4번 반복을 통해 그래프 생성
    ax = axes[i//2, i%2] # 2로 나누어서 0 혹은 1이 나오면 i가 0이면 0,0, i가 1이면 0,1...이런식
    df = kakao.loc[(kakao.index > ts[i][0]) & (kakao.index <ts[i][1])] # 
    sns.lineplot(y=df['Close'], x=df.index, ax=ax)
    ax.set_title(f'{ts[i][0]}~{ts[i][1]}')
    ax.set_xlabel('time')
    ax.set_ylabel('price')
plt.tight_layout() # 가독성을 좋게 하기 위해서
```
